```python
# Importing all the required libraries
import numpy as np
import pandas as pd
import random as rand
import math
from statistics import mean
import statistics
import matplotlib.pyplot as plt
```

```python
n = 4
A = -50
loc_A = (0,30,3)
loc_B = (50,60,4)
loc_C = (100,30,3)
```

```python

def dist(loc_1,loc_2):
    x1 = loc_1[0]
    y1 = loc_1[1]
    z1 = loc_1[2]
    x2 = loc_2[0]
    y2 = loc_2[1]
    z2 = loc_2[2]
    x = (x1 - x2)**2
    y = (y1 - y2)**2
    z = (z1 - z2)**2
    dist = math.sqrt(x+y+z)
    return round(dist,4)
```

```python
def RSSI(dist):
    rssi = (-10*n*math.log(dist))+A
    return round(rssi, 3)
```

```python
RSSI_profile=[]
d= list(range(1,141))
for i in range (1,141):
    RSSI_profile.append(RSSI(i))
RSSI_profile=np.array(RSSI_profile)
print(RSSI_profile)
print("Mean: ",mean(RSSI_profile))
print("Standard deviation: ",statistics.pstdev(RSSI_profile))
#Plotting the RSSI profile
plt.plot(d,RSSI_profile)
plt.title("RSSI Profile as a function of distance")
plt.xlabel("Distance")
plt.ylabel("RSSI Profile")
plt.plot(d,RSSI_profile,color='Green')
```

```python
def fprint(loc, loc_A, loc_B, loc_C):
    dist_A= dist(loc, loc_A)
    dist_B= dist(loc, loc_B)
    dist_C= dist(loc, loc_C)
    fprint_A= RSSI_rndm_noise(dist_A)
    fprint_B= RSSI_rndm_noise(dist_B)
    fprint_C= RSSI_rndm_noise(dist_C)
    return [fprint_A, fprint_B, fprint_C]
```

```python
def RSSI_rndm_noise(dist):
    x = np.random.normal(0,5.1,1)
    rssi = (-10*n*math.log(dist))+A+x[0]
    return round(rssi, 4)
```

```python
max_center_length = 100
max_center_width = 60
min_center_length = 0.5
min_center_width = 0.5
tiles_center =[]
for i in np.arange(min_center_length, max_center_length,1):
    for j in np.arange(min_center_width, max_center_width,1):
        tiles_center.append([i,j,0])
for i in range(len(tiles_center)):
    print("Tile at position ",i," is ",tiles_center[i])
fprint_tile_grid = []
for i in range(len(tiles_center)):
    fprint_tile_grid.append(fprint(tiles_center[i],loc_A, loc_B, loc_C))

fprint_array = np.array(fprint_tile_grid)
fprint_dataframe = pd.DataFrame(fprint_array)
#Save to csv
fprint_dataframe.to_csv("fprint_grid.csv")
```

```python
def RSSI_roam_devices(roam_rssi,fprint_array):
    # Assumption: Initially the roam device tile is the first tile index
    roam_index_best_match = 0
    # Assuming: Initially the roam device RSSI matches the first tile RSSI
    roam_rssi_best_match = [fprint_array[0]]
    # Difference between fprint of the tile and the roam_rssi should be the least possible value
    rssi_best_match_A = roam_rssi[0]-fprint_array[0][0]
    rssi_best_match_B = roam_rssi[1]-fprint_array[0][1]
    rssi_best_match_C = roam_rssi[2]-fprint_array[0][2]
    rssi_best_match_difference= math.sqrt(abs(rssi_best_match_A**2+rssi_best_match_B**2+rssi_best_match_C**2))
    for i in range(len(fprint_array)):
        rssi_A = roam_rssi[0]-fprint_array[i][0]
        rssi_B = roam_rssi[1]-fprint_array[i][1]
        rssi_C = roam_rssi[2]-fprint_array[i][2]
        rssi_difference = math.sqrt(abs(rssi_A**2+rssi_B**2+rssi_C**2))
        if(rssi_best_match_difference>rssi_difference):
            roam_index_best_match = i
            roam_rssi_best_match = [fprint_array[i]]
            rssi_best_match_A = rssi_A
            rssi_best_match_B = rssi_B
            rssi_best_match_C = rssi_C
            rssi_best_match_difference = math.sqrt(abs(rssi_best_match_A+rssi_best_match_B+rssi_best_match_C))
    return roam_index_best_match, roam_rssi_best_match
```

```python
def loc_err(true_loc, mean_loc):
    loc_err_A = mean_loc[0]-true_loc[0]
    loc_err_B = mean_loc[1]-true_loc[1]
    loc_err_C = mean_loc[2]-true_loc[2]
    loc_err_mean = math.sqrt(loc_err_A**2 + loc_err_B**2 + loc_err_C**2)
    return loc_err_mean

```

```python
def triangultion_dist(rssi_roam):
    dist_from_deviceA = math.exp((A-rssi_roam[0])/(10*n))
    dist_from_deviceB = math.exp((A-rssi_roam[1])/(10*n))
    dist_from_deviceC = math.exp((A-rssi_roam[2])/(10*n))
    return dist_from_deviceA,dist_from_deviceB, dist_from_deviceC
```

```python

def trilateration(loc_A, loc_B, loc_C, r1, r2, r3):
    loc_1 = np.array([0,0,0])
    loc_2 = np.array([loc_B[0]-loc_A[0], loc_B[1]-loc_A[1], loc_B[2]-loc_A[2]])
    loc_3 = np.array([loc_C[0]-loc_A[0], loc_C[1]-loc_A[1], loc_C[2]-loc_A[2]])
    dist_1 = loc_2 - loc_1
    dist_2 = loc_3 - loc_1
    cross_product = np.cross(dist_1, dist_2)
    x_n = (dist_1)/np.linalg.norm(dist_1)
    z_n = (cross_product)/np.linalg.norm(cross_product)
    y_n = np.cross(x_n, z_n)
    i = np.dot(x_n, dist_2)
    j = np.dot(y_n, dist_2)
    d = np.dot(x_n, dist_1)
    X = ((r1**2) - (r2**2) + (d**2))/(2*d)
    Y = (((r1**2) - (r3**2) + (i**2) + (j**2))/(2*j)) - ((i/j)*(X))
    Z = np.sqrt(max(0, r1**2 - X**2 -Y**2))
    K = loc_A + X*x_n + Y*y_n + Z*z_n
    return abs(K)
```
